```python

KIND = 'albums'
task_key = f"BRONZE_{KIND.upper()}"
```

```python
#Fetch job values between tasks

ids_to_call = dbutils.jobs.taskValues.get(taskKey = task_key, key = f"new_{KIND}_ids_to_call", debugValue = [])

ids_written = dbutils.jobs.taskValues.get(taskKey = task_key, key = f"new_{KIND}_ids_written", debugValue = [])
```

```python
# TEST
# Make sure that the values are lists
assert isinstance(ids_to_call, list),"ids_to_call is not a list"
assert isinstance(ids_written, list),"ids_written is not a list"
```

```python
# Testing we have written the same number of ids
assert len(ids_to_call) == len(ids_written),"ids_to_call and ids_written are not the same"
```

```python
# TEST 
# there are ids called not wirtten
not_written = [x for x in ids_to_call if x not in ids_written ]

assert len(not_written) == 0,"There are ids called not written"    
```
