```python
from pyspark.sql import functions as F
```

```python
sample_data = {
  "artists": "s3://my-raw-spotify-data/bronze/artists/2025-11-10/2025-11-100.json",
  "albums" : "s3://my-raw-spotify-data/bronze/albums/2025-11-08_0.json",
  "tracks" : "s3://my-raw-spotify-data/bronze/tracks/2025-11-10/2025-11-100.json"
    
}
```

ARTISTS_dev_section

```python
test_path = sample_data["tracks"]
df = (spark.read.format("json")
      .option("multiLine", "true")
      .load(test_path)
)

df.printSchema()     
```

```python
# Eploding the array of tracks

explode_df = df.select(F.explode("tracks").alias("tracks"))

```

```python
# Flattening tracks dataframe to get the columns
# No need to cast anything since the schema is correclty infered by Spark

flat_df = explode_df.select(
    F.col("tracks.id").alias("id"),
    F.col("tracks.name").alias("name"),
    F.col("tracks.duration_ms").alias("duration_ms"),
    F.col("tracks.popularity").alias("popularity"),
    F.col("tracks.explicit").alias("explicit"),
    F.col("tracks.track_number").alias("track_number"))
```
