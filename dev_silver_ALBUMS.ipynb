```python
from pyspark.sql import functions as F
```

```python
sample_data = {
  "artists": "s3://my-raw-spotify-data/bronze/artists/2025-11-10/2025-11-100.json",
  "albums" : "s3://my-raw-spotify-data/bronze/albums/2025-11-10/2025-11-100.json",
  "tracks" : "s3://my-raw-spotify-data/bronze/tracks/2025-11-10/2025-11-100.json"
    
}
```

ARTISTS_dev_section

```python
test_path = sample_data["albums"]
df = (spark.read.format("json")
      .option("multiLine", "true")
      .load(test_path)
)

df.printSchema()     
```

```python
# Eploding the array of artists

explode_df = df.select(F.explode("albums").alias("albums"))

```

```python
display(explode_df.limit(2))
```

```python
# Flattening artists dataframe to get the columns
# No need to cast anything since the schema is correclty infered by Spark

flat_df = explode_df.select(
    F.col("albums.id").alias("id"),
    F.col("albums.name").alias("name"),
    F.col("albums.type").alias("type"),
    F.col("albums.label").alias("label"),
    F.col("albums.popularity").alias("popularity"),
    F.col("albums.total_tracks").alias("total_tracks"),
    F.col("albums.release_date").alias("release_date"),
    F.col("albums.release_date_precision").alias("release_date_precision"),
    F.col("albums.images")[1]["url"]).alias("image_url_medium")
```
